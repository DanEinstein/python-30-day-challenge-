```python
# Covering the first part of EDA 
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 

data = {
    'customers': ['kibe', 'Muiruri', 'Kinya', 'wamalawa', 'Achieng', 'Otieno'],
    'sales':     [240, np.nan, 650, 450, np.nan, 300],      # np.nan in positions 1 and 4
    'Revenue':   [200, 300, np.nan, 400, 250, np.nan]        # np.nan in positions 2 and 5
}
df = pd.DataFrame(data)
```

```python
#  finding the null elements from the dataset 
df.isnull() 


```

```python
df.isnull().sum() #calculates the total number of null cells in each of the columns 
```

```python
df.isnull().sum().sum()  # total missing values in whole df
```

```python
df['sales'].isnull().sum()
```

```python
#  dropping the null elements in the dataframe 
df.dropna()
#  dropping any row with any missing value

```

```python
df.dropna(subset=['sales'])      # drop rows missing in 'sales' only
```

```python
df.dropna(axis=1)                # drop columns instead of rows
```

```python
df.dropna(thresh=2)              # keep rows with at least 2 non-null values
```

```python
#  proceeding with filling NAN columns 
#  filling with the zero value 
df.fillna(0)
```

```python
#  filling the sales and revenue column with the average of each column
df['sales'].fillna(df['sales'].mean())
df['Revenue'].fillna(df['Revenue'].mean())
```

```python
# . Interpolation — estimate missing values from surrounding data

# This is smarter than a flat fill — it guesses based on the pattern of nearby values.
df['sales'].interpolate()
print(df)
# Essence: Interpolation assumes your data has a trend or continuity — good for time series, sensor readings, prices over time. Bad for categorical or "jumpy" data where there's no real relationship between adjacent rows (like your customers column — interpolating between two names makes no sense).
```

```python
df['Revenue'].ffill()   # forward fill: carry last known value forward
```

```python
df['Revenue'].bfill() #backward fill : borrow the next known value
```
