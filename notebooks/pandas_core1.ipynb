```python
#  finally i am covering the concepts of the boolean filttering as well as the iloc and loc methods 
import pandas as pd
import numpy as np

# Load the data
df = pd.read_csv('../data/students.csv')

print("="*70)
print("ORIGINAL DATAFRAME")
print("="*70)
print(df)
print(f"\nShape: {df.shape}")
print(f"Columns: {list(df.columns)}")
print(f"Index: {list(df.index)}")
print("\n")

# ============================================================
# 1. UNDERSTANDING THE DIFFERENCE: LOC vs ILOC
# ============================================================
print("="*70)
print("1. LOC vs ILOC - THE DIFFERENCE")
print("="*70)

print("\n--- LOC uses LABELS (column names and index labels) ---")
print("df.loc[0, 'Name']  # row label 0, column label 'Name'")
print(f"Result: {df.loc[0, 'Name']}")

print("\n--- ILOC uses POSITIONS (integer positions, like lists) ---")
print("df.iloc[0, 0]  # position 0 (row), position 0 (column)")
print(f"Result: {df.iloc[0, 0]}")

print("\nBoth give the same answer, but use different methods!")
print("\n")

# ============================================================
# 2. SELECTING SINGLE CELLS
# ============================================================
print("="*70)
print("2. SELECTING SINGLE CELLS")
print("="*70)

print("\n--- Using LOC (label-based) ---")
print("df.loc[2, 'City']  # Row with label 2, column 'City'")
print(f"Result: {df.loc[2, 'City']}")

print("\n--- Using ILOC (position-based) ---")
print("df.iloc[2, 2]  # Position 2 (3rd row), position 2 (3rd column)")
print(f"Result: {df.iloc[2, 2]}")

print("\nNote: Position 2 = 3rd row/column (0-indexed)")
print("\n")

# ============================================================
# 3. SELECTING ENTIRE ROWS
# ============================================================
print("="*70)
print("3. SELECTING ENTIRE ROWS")
print("="*70)

print("\n--- Using LOC (by label) ---")
print("df.loc[1]  # Row with label 1")
print(df.loc[1])

print("\n--- Using ILOC (by position) ---")
print("df.iloc[1]  # Row at position 1 (2nd row)")
print(df.iloc[1])

print("\nBoth return the same row, different methods")
print("\n")

# ============================================================
# 4. SELECTING MULTIPLE ROWS
# ============================================================
print("="*70)
print("4. SELECTING MULTIPLE ROWS")
print("="*70)

print("\n--- Using LOC (INCLUSIVE of end) ---")
print("df.loc[0:2]  # Rows 0, 1, 2 (INCLUDES row 2)")
print(df.loc[0:2])

print("\n--- Using ILOC (EXCLUSIVE of end) ---")
print("df.iloc[0:2]  # Positions 0, 1 (EXCLUDES position 2)")
print(df.iloc[0:2])

print("\n⚠️  KEY DIFFERENCE: loc includes end, iloc excludes end!")
print("\n")

# ============================================================
# 5. SELECTING COLUMNS
# ============================================================
print("="*70)
print("5. SELECTING COLUMNS")
print("="*70)

print("\n--- Using LOC (by column name) ---")
print("df.loc[:, ['Name', 'Age']]  # All rows, specific columns")
print(df.loc[:, ['Name', 'Age']])

print("\n--- Using ILOC (by column position) ---")
print("df.iloc[:, [0, 1]]  # All rows, positions 0 and 1")
print(df.iloc[:, [0, 1]])

print("\n")

# ============================================================
# 6. SELECTING ROWS AND COLUMNS TOGETHER
# ============================================================
print("="*70)
print("6. SELECTING ROWS AND COLUMNS TOGETHER")
print("="*70)

print("\n--- Using LOC (labels) ---")
print("df.loc[0:3, ['Name', 'City']]  # Rows 0-3, Name and City columns")
print(df.loc[0:3, ['Name', 'City']])

print("\n--- Using ILOC (positions) ---")
print("df.iloc[0:3, [0, 2]]  # Rows 0-2 (pos), columns 0,2 (pos)")
print(df.iloc[0:3, [0, 2]])

print("\n")

# ============================================================
# 7. BOOLEAN FILTERING - SINGLE CONDITIONS
# ============================================================
print("="*70)
print("7. BOOLEAN FILTERING - SINGLE CONDITIONS")
print("="*70)

print("\n--- Filter: Age > 25 ---")
print("df[df['Age'] > 25]")
result = df[df['Age'] > 25]
print(result)

print("\n--- Filter: City == 'Nairobi' ---")
print("df[df['City'] == 'Nairobi']")
result = df[df['City'] == 'Nairobi']
print(result)

print("\n--- Filter: Grade == 'A' ---")
print("df[df['Grade'] == 'A']")
result = df[df['Grade'] == 'A']
print(result)

print("\n")

# ============================================================
# 8. BOOLEAN FILTERING - MULTIPLE CONDITIONS (AND)
# ============================================================
print("="*70)
print("8. BOOLEAN FILTERING - MULTIPLE CONDITIONS (AND)")
print("="*70)

print("\n--- Filter: Age > 25 AND City == 'Nairobi' ---")
print("df[(df['Age'] > 25) & (df['City'] == 'Nairobi')]")
result = df[(df['Age'] > 25) & (df['City'] == 'Nairobi')]
print(result)

print("\n--- Filter: Grade == 'A' AND Age < 30 ---")
print("df[(df['Grade'] == 'A') & (df['Age'] < 30)]")
result = df[(df['Grade'] == 'A') & (df['Age'] < 30)]
print(result)

print("\n⚠️  IMPORTANT: Use & for AND, not 'and' keyword!")
print("\n")

# ============================================================
# 9. BOOLEAN FILTERING - MULTIPLE CONDITIONS (OR)
# ============================================================
print("="*70)
print("9. BOOLEAN FILTERING - MULTIPLE CONDITIONS (OR)")
print("="*70)

print("\n--- Filter: Grade == 'A' OR Age > 40 ---")
print("df[(df['Grade'] == 'A') | (df['Age'] > 40)]")
result = df[(df['Grade'] == 'A') | (df['Age'] > 40)]
print(result)

print("\n--- Filter: City == 'Nairobi' OR City == 'Kisumu' ---")
print("df[(df['City'] == 'Nairobi') | (df['City'] == 'Kisumu')]")
result = df[(df['City'] == 'Nairobi') | (df['City'] == 'Kisumu')]
print(result)

print("\n⚠️  IMPORTANT: Use | for OR, not 'or' keyword!")
print("\n")

# ============================================================
# 10. BOOLEAN FILTERING - NOT CONDITIONS
# ============================================================
print("="*70)
print("10. BOOLEAN FILTERING - NOT CONDITIONS")
print("="*70)

print("\n--- Filter: NOT Grade == 'A' (i.e., Grade != 'A') ---")
print("df[~(df['Grade'] == 'A')]")
result = df[~(df['Grade'] == 'A')]
print(result)

print("\n--- Filter: NOT from Mombasa ---")
print("df[~(df['City'] == 'Mombasa')]")
result = df[~(df['City'] == 'Mombasa')]
print(result)

print("\n⚠️  IMPORTANT: Use ~ for NOT!")
print("\n")

# ============================================================
# 11. BOOLEAN FILTERING - USING .isin()
# ============================================================
print("="*70)
print("11. BOOLEAN FILTERING - USING .isin()")
print("="*70)

print("\n--- Filter: Students from Nairobi OR Kisumu ---")
print("cities = ['Nairobi', 'Kisumu']")
print("df[df['City'].isin(cities)]")
cities = ['Nairobi', 'Kisumu']
result = df[df['City'].isin(cities)]
print(result)

print("\n--- Filter: Students with Grade A OR B ---")
print("grades = ['A', 'B']")
print("df[df['Grade'].isin(grades)]")
grades = ['A', 'B']
result = df[df['Grade'].isin(grades)]
print(result)

print("\n✓ .isin() is cleaner for multiple values!")
print("\n")

# ============================================================
# 12. COMBINING LOC/ILOC WITH BOOLEAN FILTERING
# ============================================================
print("="*70)
print("12. COMBINING LOC/ILOC WITH BOOLEAN FILTERING")
print("="*70)

print("\n--- Get Name and Age of students with Age > 25 ---")
print("df.loc[df['Age'] > 25, ['Name', 'Age']]")
result = df.loc[df['Age'] > 25, ['Name', 'Age']]
print(result)

print("\n--- Get first 3 columns of students from Nairobi ---")
print("df.loc[df['City'] == 'Nairobi', :]  or  df.loc[df['City'] == 'Nairobi']")
result = df.loc[df['City'] == 'Nairobi']
print(result)

print("\n--- Get specific columns for filtered rows ---")
print("df.loc[(df['Age'] > 25) & (df['Grade'] == 'A'), ['Name', 'City']]")
result = df.loc[(df['Age'] > 25) & (df['Grade'] == 'A'), ['Name', 'City']]
print(result)

print("\n")

# ============================================================
# 13. STRING OPERATIONS
# ============================================================
print("="*70)
print("13. STRING FILTERING (.str methods)")
print("="*70)

print("\n--- Filter: Names containing 'a' ---")
print("df[df['Name'].str.contains('a')]")
result = df[df['Name'].str.contains('a')]
print(result)

print("\n--- Filter: Names starting with 'J' ---")
print("df[df['Name'].str.startswith('J')]")
result = df[df['Name'].str.startswith('J')]
print(result)

print("\n--- Filter: Names ending with 'n' ---")
print("df[df['Name'].str.endswith('n')]")
result = df[df['Name'].str.endswith('n')]
print(result)

print("\n")

# ============================================================
# 14. PRACTICAL EXAMPLES
# ============================================================
print("="*70)
print("14. PRACTICAL EXAMPLES")
print("="*70)

print("\n--- Example 1: Get Nairobi students with Grade A ---")
result = df.loc[(df['City'] == 'Nairobi') & (df['Grade'] == 'A'), 'Name']
print(f"Names: {list(result)}")

print("\n--- Example 2: Count students per city ---")
print(df['City'].value_counts())

print("\n--- Example 3: Average age by grade ---")
print(df.groupby('Grade')['Age'].mean())

print("\n--- Example 4: Get oldest student ---")
oldest = df.loc[df['Age'].idxmax()]
print(f"{oldest['Name']} is {oldest['Age']} years old")

print("\n--- Example 5: Get students in age range 25-35 ---")
result = df[(df['Age'] >= 25) & (df['Age'] <= 35)]
print(result)

print("\n")

# ============================================================
# 15. QUICK REFERENCE TABLE
# ============================================================
print("="*70)
print("15. QUICK REFERENCE")
print("="*70)

reference = """
┌─────────────────────────────────────────────────────────────────┐
│                   LOC vs ILOC CHEATSHEET                        │
├─────────────────────────────────────────────────────────────────┤
│ Task                    │ LOC             │ ILOC                │
├─────────────────────────────────────────────────────────────────┤
│ Single cell             │ loc[0, 'Name']  │ iloc[0, 0]          │
│ Single row              │ loc[0]          │ iloc[0]             │
│ Rows 0-2 (inclusive)    │ loc[0:2]        │ iloc[0:3]           │
│ Single column           │ loc[:, 'Name']  │ iloc[:, 0]          │
│ Multiple columns        │ loc[:, cols]    │ iloc[:, [0, 2]]     │
│ Rows + Columns          │ loc[0:2, cols]  │ iloc[0:2, 0:2]      │
├─────────────────────────────────────────────────────────────────┤
│ Boolean filter + get    │ loc[condition,  │ Can't combine       │
│ specific columns        │  cols]          │ (use loc instead)   │
└─────────────────────────────────────────────────────────────────┘

BOOLEAN OPERATORS:
  AND  → &   (not 'and')
  OR   → |   (not 'or')
  NOT  → ~   (not 'not')

ALWAYS use parentheses around conditions:
  ✓ df[(df['Age'] > 25) & (df['City'] == 'Nairobi')]
  ✗ df[df['Age'] > 25 & df['City'] == 'Nairobi']
"""

print(reference)
```
