```python
#  Multindexing which allows you o deal with higher dimension data in a 2D frame 
import pandas as pd
import numpy as np

# Sample data: sales for different stores across years/quarters
data = {
    'store': ['NYC', 'NYC', 'NYC', 'NYC', 'LA', 'LA', 'LA', 'LA'],
    'year': [2023, 2023, 2024, 2024, 2023, 2023, 2024, 2024],
    'quarter': ['Q1', 'Q2', 'Q1', 'Q2', 'Q1', 'Q2', 'Q1', 'Q2'],
    'revenue': [50000, 55000, 60000, 62000, 40000, 43000, 45000, 48000],
    'units_sold': [500, 550, 600, 620, 400, 430, 450, 480]
}

df = pd.DataFrame(data)

# Build a 3-level MultiIndex: store > year > quarter
df = df.set_index(['store', 'year', 'quarter']).sort_index()

print(df)
```

```python
# 1. Grab all data for one store
df.loc['NYC']

# 2. Grab a specific store/year combo
df.loc[('NYC', 2023)]

# 3. Drill down to one exact record
df.loc[('NYC', 2023, 'Q1')]

# 4. Get all Q1 data across every store/year (cross-section)
df.xs('Q1', level='quarter')

# 5. Aggregate by any level — e.g. total revenue per store
df.groupby(level='store')['revenue'].sum()
# LA     176000
# NYC    227000

# 6. Year-over-year comparison per store — reshape with unstack
df['revenue'].unstack(level='quarter')
#                Q1     Q2
# store year               
# LA    2023   40000  43000
#       2024   45000  48000
# NYC   2023   50000  55000
#       2024   60000  62000

# 7. Rolling average within each store (common combo: MultiIndex + rolling)
df.groupby(level='store')['revenue'].rolling(2).mean()
```

```python
#  Rolling window functions 
data = {
    "customers":['kibe','Muiruri','Kinya','wamalawa'],
    'sales':[240,600,650,450],
    'Revenue':[200,300,150,400]
}
df = pd.DataFrame(data)
df['rolling_avg'] = df['sales'].rolling(window=2,min_periods=1).mean()
df['rolling_sum'] = df['sales'].rolling(2).sum()
df['rolling_std'] = df['sales'].rolling(2).std()
df['rolling_max'] = df['sales'].rolling(2).max()
print(df)
```

```python
# Quick Reference Summary
# Use Case	Pandas Method	SQL Equivalent
# Moving Average (fixed size)	.rolling(window=N).mean()	AVG(...) OVER (ROWS N PRECEDING)
# Cumulative Sum	.expanding().sum()	SUM(...) OVER (ROWS UNBOUNDED PRECEDING)
# Exponential Moving Avg	.ewm(span=N).mean()	n/a (proprietary)
# Rank within a group	.groupby('cat')['col'].rank()	RANK() OVER (PARTITION BY cat)
# Row number within a group	.groupby('cat').cumcount()	ROW_NUMBER() OVER (PARTITION BY cat)
# Partitioned aggregate	.groupby('cat')['col'].transform('sum')	SUM(...) OVER (PARTITION BY cat)
# Partitioned Moving Avg	.groupby('cat')['col'].rolling(N).mean()	AVG(...) OVER (PARTITION BY cat ROWS N PRECEDING)
# Previous/Next row	.shift(1) / .shift(-1)	LAG() / LEAD()
```
