```python
#  pivot tables and melt 
sales_data ={
    'salesperson':['Alice','Alice','Bob','Bob','charles','charles'],
    'month':['january','february','january','february','january','february'],
    'sales':[600,500,550,400,650,700]
}
df_sales =pd.DataFrame(sales_data)
print(df_sales)

#  The above is the data that we  will use to show pivot tables
```

```python
pivot_table = df_sales.pivot( index='salesperson', columns='month', values='sales')
print(pivot_table)
#  This is the simple implementation of the pivot tables
```

```python
# Implementation of pivot tables and aggregate functions 
data_pivot_table = df_sales.pivot_table(
    index='salesperson', 
    columns='month', 
    values='sales',
    aggfunc='sum'  # Combine duplicates by summing
)
print(data_pivot_table)
#  other aggregate functions include aggfunc(max), aggfunc(min), aggfunc(mean), aggfunc(count)

```

```python
#  Here we are going to explore melt 
# 3. MELT - Reshaping Data (Long Format)

# Melt is the opposite of pivot - transforms from wide format → long format
melted = pivot_table.reset_index().melt(
    id_vars='salesperson',  # Note: Capital S!
    var_name='month',
    value_name='sales'
)
print(melted)
```

```python
#  working with a complete examples that covers both pivot and melt 
import pandas as pd

# Example 1: Simple Pivot
df = pd.DataFrame({
    'Salesperson': ['Alice', 'Alice', 'Bob', 'Bob', 'Charlie', 'Charlie'],
    'Month': ['January', 'February', 'January', 'February', 'January', 'February'],
    'Sales': [500, 600, 400, 550, 700, 650]
})

# Create pivot table
pivot = df.pivot(index='Salesperson', columns='Month', values='Sales')
print("PIVOT RESULT:")
print(pivot)
print("\n")

# Example 2: Melt the pivoted table back
melted = pivot.reset_index().melt(id_vars='Salesperson', var_name='Month', value_name='Sales')
print("MELT RESULT:")
print(melted)
print("\n")

# Example 3: Pivot with aggregation
df_duplicate = pd.DataFrame({
    'Salesperson': ['Alice', 'Alice', 'Alice', 'Bob', 'Bob', 'Bob'],
    'Month': ['January', 'January', 'February', 'January', 'January', 'February'],
    'Sales': [500, 300, 600, 400, 200, 550]  # Duplicates!
})

pivot_agg = df_duplicate.pivot_table(
    index='Salesperson',
    columns='Month',
    values='Sales',
    aggfunc='sum'  # Combine duplicates
)
print("PIVOT WITH AGGREGATION:")
print(pivot_agg)
print("\n")

# Example 4: Student grades example
df_grades = pd.DataFrame({
    'Student': ['Alice', 'Alice', 'Bob', 'Bob', 'Charlie', 'Charlie'],
    'Subject': ['Math', 'English', 'Math', 'English', 'Math', 'English'],
    'Score': [85, 90, 92, 88, 78, 95],
    'Grade': ['A', 'A', 'A+', 'B', 'B', 'A']
})

# Pivot scores
pivot_scores = df_grades.pivot(index='Student', columns='Subject', values='Score')
print("STUDENT SCORES (Pivoted):")
print(pivot_scores)
print("\n")

# Melt back
melted_grades = df_grades.melt(id_vars='Student', 
                                var_name='Test_Type',
                                value_name='Result')
print("MELTED GRADES:")
print(melted_grades)
```

```python
#  creating and working with Datetimes 
import pandas as pd
from datetime import datetime, timedelta
print('imports successfully')
#   using the pd.date_time() on strings
dates = pd.to_datetime(['2024-01-15', '2024-02-20', '2024-03-10'])
print(dates)

#  using the datetime.date objects 
dates = pd.to_datetime([datetime(2024,1,15),datetime(2024,2,16),datetime(2026,1,19)])
print(dates)
#  creating a date range 
dates = pd.date_range(start = '2025-01-01', end = '2025-01-25', freq = 'D')
print(dates)

#  method 4 creating the date range using the different frequencies 
daily = pd.date_range('2024-01-01', periods=5, freq='D')    # Daily
weekly = pd.date_range('2024-01-01', periods=5, freq='W')   # Weekly
monthly = pd.date_range('2024-01-01', periods=5, freq='MS')  # Monthly

print(daily)
print(weekly)
print(monthly)
```

```python
#  parsing different date formats 
dates = pd.to_datetime(['01/15/2024','01/16/2024','01/17/2024'],format='%m/%d/%Y')
print(dates)
# European format
dates = pd.to_datetime(['15-01-2024', '20-02-2024'], format='%d-%m-%Y')
print(dates)
#  with time 
datetime = pd.to_datetime(['2024-09-15 10:30:45','2024-10-16 14:15:20'])
print(datetime)

#  using the mixed type 
dates = pd.to_datetime(['2024/09/16','2025-06-09'], format = 'mixed')
print(dates)
# The mixed format is normally slow as compared to the rest 

```

```python
#  Time series Analysis
import pandas as pd
import numpy as np

# Create sample sales data
dates = pd.date_range('2024-01-01', periods=30, freq='D')
data = {
    'Date': dates,
    'Sales': np.random.randint(100, 200, 30),
    'Product': np.random.choice(['A', 'B', 'C'], 30)
}

df = pd.DataFrame(data)

# Extract date components
df['Month'] = df['Date'].dt.month
df['Week'] = df['Date'].dt.isocalendar().week
df['DayName'] = df['Date'].dt.day_name()

# Calculate statistics
df['7Day_MA'] = df['Sales'].rolling(window=7).mean()
df['Daily_Change'] = df['Sales'].diff()

# Set date as index
df_indexed = df.set_index('Date')

# Resample by week
weekly_sales = df_indexed['Sales'].resample('W').sum()

# Resample by month
monthly_sales = df_indexed['Sales'].resample('ME').mean()

# Filter to January data
january = df[df['Date'].dt.month == 1]

# Group by product and get monthly stats
product_stats = df.groupby('Product').resample('W', on='Date')['Sales'].sum()

print("DataFrame head:")
print(df.head())
print("\nWeekly Sales:")
print(weekly_sales)
print("\nMonthly Average Sales:")
print(monthly_sales)
```
