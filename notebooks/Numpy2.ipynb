```python
# solving systems of linear equations 
# example is solving:
# 1x0 + 2x1 = 1
# 3x0 + 5x1 = 2
# we will implement the above to solve the equations 
A = np.array([[1,2],[3,5]])
b = np.array([1,2])
x = np.linalg.solve(A,b)
print(x)
```

```python
#  Random shuffles and permutations
# when using the shuffke methods it modifies the particular array 
# The permutations method create a copy of the array and works with that 
arr = np.array([1,2,3,4,5])
rng.shuffle(arr)
print(arr)
```

```python
# using the permutations method now 
arr = np.array([1,2,3,4,5])
new_arr = rng.permutation(arr)
print(arr)
print(new_arr)
# The old array is left intact when the new array is being modified 
```

```python
#  The binomial distributions 
rng = np.random.default_rng()
n, p = 10, .5  # number of trials, probability of each trial
s = rng.binomial(n, p, 1000)
#  This presents the number of trials of flipping a coin 10 times and testing the same in 1000 trials 
```

```python
#   A company drills 9 wild-cat oil exploration wells, each with an estimated probability of success of 0.1. All nine wells fail. What is the probability of that happening?

# Let’s do 20,000 trials of the modiel, and count the number that generate zero positive results.

result = sum(rng.binomial(9, 0.1, 20000) == 0)/20000.
print(result) 
#  about 39 percent 


```

```python
#  Testing the ransom distribution 
mu, sigma = 0,0.1 
s = np.random.default_rng().normal(mu,sigma,1000)
# verifying the mean

abs(mu - np.mean(s))
0.0  # may vary
abs(sigma - np.std(s, ddof=1))
# verifying the standard deviation and it may vary as well
import matplotlib.pyplot as plt
count, bins, ignored = plt.hist(s, 30, density=True)
#  plotting the normal distribution 
plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) *
               np.exp( - (bins - mu)**2 / (2 * sigma**2) ),
         linewidth=2, color='y')
plt.show()
#  it is important to note that the values produced are random 

```
