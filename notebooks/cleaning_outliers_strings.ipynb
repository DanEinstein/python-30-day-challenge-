```python
# using the outliers 
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 

data = {
    'customers': ['kibe', 'Muiruri', 'Kinya', 'wamalawa', 'Achieng', 'Otieno', 'Wafula'],
    'sales':     [240, np.nan, 650, 450, np.nan, 300,9800],      # np.nan in positions 1 and 4
    'Revenue':   [200, 300, np.nan, 400, 250, np.nan,np.nan]        # np.nan in positions 2 and 5
}
df = pd.DataFrame(data)
```

```python
#  IQR(inter-quartile range)
Q1 = df['sales'].quantile(0.25)
Q3 = df['sales'].quantile(0.75)
IQR = Q3-Q1


lower_bound = Q1 - 1.5*IQR
upper_bound = Q3 + 1.5*IQR

outliers = df[(df['sales']>upper_bound)|(df['sales']<lower_bound)]
print(outliers)
```

```python
#  using the z-score 
mean = df['sales'].mean()
std = df['sales'].std()
df['z_score'] = (df['sales'] - mean) / std

outliers = df[df['z_score'].abs() > 3]
print(outliers)
```

```python
#  visual inspection 
df.boxplot(column='sales')          # quick one-liner in pandas
df['sales'].hist(bins=10)           # distribution shape
df['sales'].plot(kind='box')        # same as boxplot, alt syntax
```

```python
#  string cleaning 
#  exploring the string cleaning 
df['customers'] = df['customers'].str.strip()
#  rstrip and left strip also help in the cleaning of the whitespaces either on the left or on the right 

```

```python
#  .str.replace() allows the replacing of certain characters 
df['customers'] = df['customers'].str.lower()   #standardization of the dataset 
df['customers'] = df['customers'].str.replace('!', '')  #replacing some chars 
df['customers'] = df['customers'].str.replace('_', ' ') 
print(df)

```
